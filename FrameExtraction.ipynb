```python
import cv2
import os
```

```python
def list_files(dir):
    r = []
    for root, dirs, files in os.walk(dir):
        for name in files:
            r.append(os.path.join(root, name))
    return r
```

```python
files = list_files("CUSTOM DATASET")
files
```

```python
for file in files:
    vid = cv2.VideoCapture(file)
    count = 0
    folder = "DATASET_FRAMES" + file[8:-4]
    os.makedirs(folder, exist_ok=True)
    while True:
        success, frame = vid.read()
        if not success:
            break
        cv2.imshow("Output", frame)
        filename = os.path.join(folder, f"frame_{count:04d}.jpg")
        cv2.imwrite(filename, frame)
        count += 1
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    vid.release()
    cv2.destroyAllWindows()
```
